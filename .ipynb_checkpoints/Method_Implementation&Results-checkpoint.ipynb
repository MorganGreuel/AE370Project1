```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constants
g = 9.81  # m/s^2
rho_0 = 1.225  # kg/m^3
H = 8500  # scale height in meters

# Minuteman III Stats (approximate)
stage = {
    'dry_mass': 10000,         # kg (stage mass without fuel)
    'fuel_mass': 26030,        # kg
    'burn_rate': 434,          # kg/s (fuel_mass / burn_time ~ 60s)
    'T': 901000,               # N (thrust)
    'theta_deg': 45,           # degrees
    'Cd': 0.2,
    'A': np.pi * (1.67 / 2)**2  # m^2 (diameter = 1.67m)
}

# Burn time
burn_time = stage['fuel_mass'] / stage['burn_rate']
m0 = stage['dry_mass'] + stage['fuel_mass']
theta = np.radians(stage['theta_deg'])

# Drag function
def drag_vectorized(vx, vy, y, Cd, A):
    v = np.sqrt(vx**2 + vy**2)
    rho = rho_0 * np.exp(-y / H)
    rho[y >= 1e5] = 0
    D = 0.5 * rho * Cd * A * v
    return D * vx, D * vy

# First-order ODE for position, velocity, and acceleration

def rocket_ode(t, z):
    x, y, vx, vy, delta_v = z
    m = m0 - stage['burn_rate'] * t if t < burn_time else stage['dry_mass']
    T_x = stage['T'] * np.cos(theta) if t < burn_time else 0
    T_y = stage['T'] * np.sin(theta) if t < burn_time else 0

    D_x, D_y = drag_vectorized(np.array([vx]), np.array([vy]), np.array([y]), stage['Cd'], stage['A'])
    D_x, D_y = D_x[0], D_y[0]

    ax = (T_x - D_x) / m
    ay = (T_y - D_y) / m - g

    return [vx, vy, ax, ay, np.sqrt(ax**2 + ay**2)]

# Initial state
z0 = [0, 0, 0, 0, 0]  # x, y, vx, vy, delta_v

t_span = (0, 200)
t_eval = np.linspace(*t_span, 1000)
sol = solve_ivp(rocket_ode, t_span, z0, t_eval=t_eval, rtol=1e-8, atol=1e-8)

# Extract data
t = sol.t
x, y = sol.y[0], sol.y[1]
vx, vy = sol.y[2], sol.y[3]
a = sol.y[4]
v = np.sqrt(vx**2 + vy**2)

D_x, D_y = drag_vectorized(vx, vy, y, stage['Cd'], stage['A'])
D = np.sqrt(D_x**2 + D_y**2)

thrust = np.where(t < burn_time, stage['T'], 0)

# Plotting
fig, axs = plt.subplots(3, 2, figsize=(14, 12))

# 1. Acceleration vs Time
axs[0, 0].plot(t, a / g)
axs[0, 0].set_title("Acceleration vs Time")
axs[0, 0].set_xlabel("Time (s)")
axs[0, 0].set_ylabel("a(t) [g]")
axs[0, 0].grid(True)

# 2. Velocity vs Time
axs[0, 1].plot(t, v / 1000)
axs[0, 1].set_title("Velocity vs Time")
axs[0, 1].set_xlabel("Time (s)")
axs[0, 1].set_ylabel("v(t) [km/s]")
axs[0, 1].grid(True)

# 3. Drag Force vs Altitude
y_km = y / 1000
D_kN = D / 1000
axs[1, 0].plot(y_km, D_kN)
axs[1, 0].set_title("Drag Force vs Altitude")
axs[1, 0].set_xlabel("Altitude (km)")
axs[1, 0].set_ylabel("D(y) [kN]")
axs[1, 0].grid(True)

# 4. Acceleration vs Altitude
a_g = a / g
axs[1, 1].plot(y_km, a_g)
axs[1, 1].set_title("Acceleration vs Altitude")
axs[1, 1].set_xlabel("Altitude (km)")
axs[1, 1].set_ylabel("a(y) [g]")
axs[1, 1].grid(True)

# 5. Thrust vs Time
axs[2, 0].plot(t, thrust / 1000)
axs[2, 0].set_title("Thrust vs Time")
axs[2, 0].set_xlabel("Time (s)")
axs[2, 0].set_ylabel("T(t) [kN]")
axs[2, 0].grid(True)

# 6. Delta-V vs Time
axs[2, 1].plot(t, v / 1000)
axs[2, 1].set_title("Delta-V vs Time")
axs[2, 1].set_xlabel("Time (s)")
axs[2, 1].set_ylabel("Δv(t) [km/s]")
axs[2, 1].grid(True)

plt.tight_layout()
plt.show()

# --- Theta-dependent Range Graph ---
def simple_range(theta_deg):
    theta = np.radians(theta_deg)
    v_burnout = (stage['T'] / m0) * burn_time
    vx = v_burnout * np.cos(theta)
    vy = v_burnout * np.sin(theta)
    time_of_flight = 2 * vy / g
    return vx * time_of_flight

angles = np.linspace(5, 85, 50)
ranges = [simple_range(angle) / 1000 for angle in angles]  # km

plt.figure(figsize=(8, 6))
plt.plot(angles, ranges)
plt.title("Range vs Launch Angle")
plt.xlabel("Launch Angle (degrees)")
plt.ylabel("Range (km)")
plt.grid(True)
plt.show()
```
